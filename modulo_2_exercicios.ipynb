## 1\. Listas

Crie uma lista chamada `filmes` com o nome dos 10 primeiros filmes mais bem avaliados no site no [IMDB](https://www.imdb.com/chart/top/). Imprima o resultado.

```
#lista solicitada no exemplo
print ('Top 10 filmes IMDB:')
print ('------------------------')

filmes = ['01 Um Sonho de Liberdade ', '02 O Poderoso Chefão', '03 Batman: O Cavaleiro das Trevas', '04 O Poderoso Chefão II', '05 12 Homens e uma Sentença', '06 A Lista de Schindler', '07 O Senhor dos Anéis: O Retorno do Rei', '08 Pulp Fiction - Tempo de Violência', '09 O Senhor dos Anéis: A Sociedade do Anel', '10 Três Homens em Conflito']
print (filmes)
```

```
#Segunda lista para estudo de lista com quebra de linhas
print ('Lista com quebra de linhas:')
print ('------------------------')

def quebra_linha(list):
    for item in list:
        print(item)

filmes_quebra = ['01 Um Sonho de Liberdade ', '02 O Poderoso Chefão', '03 Batman: O Cavaleiro das Trevas', '04 O Poderoso Chefão II', '05 12 Homens e uma Sentença', '06 A Lista de Schindler', '07 O Senhor dos Anéis: O Retorno do Rei', '08 Pulp Fiction - Tempo de Violência', '09 O Senhor dos Anéis: A Sociedade do Anel', '10 Três Homens em Conflito']

quebra_linha (filmes_quebra)
```

Simule a movimentação do *ranking*. Utilize os métodos `insert` e `pop` para trocar a posição do primeiro e do segundo filme da lista. Imprima o resultado.



```
#metodo pop
nova_posicao = filmes.pop(0)
print(nova_posicao)
nova_posicao = filmes.pop(0)
print(nova_posicao)

print(filmes)
```

```
#Metodo Insert
filmes.insert(0,'01 O Poderoso Chefão' )
filmes.insert (1,'02 Um Sonho de Liberdade' )
print(filmes)
```

## 2\. Conjuntos

Aconteceu um erro no seu *ranking*. Simule a duplicação dos três últimos filmes da lista. Imprima o resultado.

```
filmes_rep = ['08 Pulp Fiction - Tempo de Violência', '09 O Senhor dos Anéis: A Sociedade do Anel', '10 Três Homens em Conflito']
print ('Filmes ultima lista:')
print (filmes)
print ('---------------')
print ('Filmes com repetição:')
filmes_novo = filmes+filmes_rep
print(filmes_novo)
```

Utiliza a conversão `set` e `list` para remover os valores duplicados. Imprima o resultado.

```
print ('Ultima lista:')
print (filmes_novo)
print ('Lista nova com set')
print(len(filmes_novo))

print('--------------------')
print('Lista corrigida')

filmes_s_rep = list(set(filmes_novo+filmes))
print (filmes_s_rep)
print('Total de filmes:')
print(len(filmes_s_rep))
```
